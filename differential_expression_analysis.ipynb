## Count per Million Normalization

The gene counts that we see here are not directly comparable as the count of the gene also depends on sequencing depth (Total number of reads collected). So here I'll be performing CPM normalization. By dividing by total reads we can make the samples comparable 

```python
def cpm_normalization(data):
    data.iloc[:,1:]= data.iloc[:,1:].div(data.iloc[:,1:].sum(axis=0),axis=1) * 1_000_000
    return data

csv_data = cpm_normalization(csv_data)
display(csv_data)
```

```python
sums_series = csv_data.iloc[:,1:].sum(axis=0)
display(sums_series)
```

Here we can see all samples sum to 1,000,000, this means they have been successfully normalized

Now I'll remove which have very low expression, here I'll set the threshold to 10, and remove any gene that have expression less than 10.

```python

def apply_cpm_threshold(data,col_name1,col_name2,threshold):

    def select_column(data,column):
        column_bool_list = data.columns == column
        column_data = csv_data.loc[:,column_bool_list]
        return column_data

    column1_data = select_column(data,col_name1)
    column2_data = select_column(data,col_name2)

    keep_gene = (column1_data.mean(axis=1) > threshold) | (column2_data.mean(axis=1)>threshold)
    data = data.loc[keep_gene,:]
    
    return data

csv_data = apply_cpm_threshold(csv_data,"UC","CD",10)
csv_data

```

## Diffrential Gene Analysis

### Kolmogorov-Smirnov Test (KS)


KS test is not commonly used in differential gene expression analysis because it does not specifically target mean differences, which are usually of biological interest. However, it is useful in a beginner setting because it is simple to implement and does not require strong assumptions.

In this project, KS test compares the distribution of gene expression values across samples (26 UC vs 22 CD) and identifies genes whose expression patterns differ between conditions.

It can detect differences beyond just the mean, such as variability, skewness, or subpopulation effects. However, for typical differential expression tasks where consistent up/down regulation is the goal, methods like DESeq2 are more appropriate and statistically powerful.

Now our first step is to divide our samples into 2 different data frames. the first data frame would be for UC and the other CD for applying the ks test

```python
def get_only_column_with_class(data,column_name):
    column_df = data.loc[:,[column_name]]
    return column_df
uc_df = get_only_column_with_class(csv_data,"UC")
cd_df = get_only_column_with_class(csv_data,"CD")
display(uc_df.shape)
display(cd_df.shape)
```

```python
from scipy.stats import ks_2samp
import numpy as np

def ks_test_to_dataframe(df1,df2):
    p_values = []
    ks_stats = []

    for i in range(df1.shape[0]):
        
        df1_values = df1.iloc[i, :].values
        df2_values = df2.iloc[i, :].values

        stat, p = ks_2samp(df1_values, df2_values)
        
        ks_stats.append(stat)
        p_values.append(p)

    p_values = np.array(p_values)
    ks_stats = np.array(ks_stats)

    return p_values,ks_stats

p_values,ks_stats = ks_test_to_dataframe(uc_df,cd_df)

```

### Implimenting log2fold

log2fold tells us the differenece between gene expression across the condition. 

It exactly tells us how expression changes between conditions

- +1 means 2 times more expressed in one condtion 
- +2 means 4 times more expressed in one condtion 
- -1 means 2 times less expressed in one condtion 
- -2 means 4 times less expressed in one condtion 

This gives a very good intuition to judge the genes 

```python
def log2fold(df1,df2):

    avoid_divide_by_zero = 1 
    log2_fold_change_list = []

    for i in range(df1.shape[0]):

        df1_value = df1.iloc[i,:].values
        df2_value = df2.iloc[i,:].values
    
        df1_mean = df1_value.mean()
        df2_mean = df2_value.mean()
    
        fold_change = (df1_mean + avoid_divide_by_zero)/(df2_mean + avoid_divide_by_zero)
        log2_fold_change = np.log2(fold_change)
    
        log2_fold_change_list.append(log2_fold_change)
        
    log2_fold_change_list = np.array(log2_fold_change_list)
    return log2_fold_change_list
    
    
log2_fc = log2fold(uc_df,cd_df)
display(log2_fc)
```

#### Applying Benjamini–Hochberg correction


As we are testing tens of thounsands of samples by chance there will be alot of false negetives. To remove these false negetives we apply Benjamini-Hochberg Correction

```python
p_values_bonf = p_values * csv_data["geneid"].shape[0]
p_values_bonf
```

#### Finding Intresting Genes

```python
### Applying filter
# Keeping where genes are unlikely to be random chance p_value_bonf <=0.5
# Keeping Genes where expression is large log_FC > 2

to_keep = (p_values_bonf <= 0.05) & (np.abs(log2_fc) >= 0.5)
print("Total genes that match your condition: ",to_keep.sum())
deg_genes = csv_data.loc[to_keep, "geneid"]
print("\nGene index and name")
display(deg_genes)
```
